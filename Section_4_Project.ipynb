```python
def load_rand(dir_path):

    X = []
    dir_path = dir_path

    for sub_dir in tqdm(os.listdir(dir_path)): # 반복문 진행율 표시를 위해 tqdm()사용한다.
        path_main = os.path.join(dir_path, sub_dir)
        
        for i, img_name in enumerate(os.listdir(path_main)):
            if i>=6: break # 6개만 임시로 불러오기
            img = cv2.imread(os.path.join(path_main,img_name)) # path_main + img_name 이미지를 불러온다.
            img = cv2.resize(img,(100,100)) # 이미지를 100x100으로 리사이징한다.
            img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # 컬러로 이미지를 불러온다.
            X.append(img)

    return X
```

```python
X = load_rand(fruit_train_dir)
```

```python
X=np.array(X) # 36개의 데이터, 100x100사이즈, RGB컬러
X.shape
```

```python
def show_subpot(X, title=False, Y=None):

    if X.shape[0] == 36:

        f, ax = plt.subplots(6, 6, figsize=(20,20)) 
        list_fruits=['freshbanana','rottenapples','freshapples','rottenbanana','freshoranges','rottenoranges']

        for i, img in enumerate(X):
            ax[i//6][i%6].imshow(img, aspect='auto') # aspect 배율을 자동으로 조정
            if title==False:
                ax[i//6][i%6].set_title(list_fruits[i//6])
            elif title and Y is not None:
                ax[i//6][i%6].set_title(Y[i])
        plt.show()
      
    else:
        print('Cannot found plot')
```

## ImageDataGenerator를 이용하여 다양한 이미지 생성

```python
datagen_train = ImageDataGenerator( 
  rotation_range=20,
  width_shift_range=0.1,
  height_shift_range=0.1,
  shear_range=0.1,
  zoom_range=0.1,
  horizontal_flip=True,
  vertical_flip=True,
  rescale=1./255,  
  validation_split=0.2,       
# validation_split: 검증을 위해 이미지를 나눈다.
# 예를들면 0.2 값을 사용하면 샘플의 20%는 검증데이터로 사용되며 나머지 80%는 훈련데이터로 사용된다.                   
  preprocessing_function=keras.applications.vgg16.preprocess_input
  )

datagen_test = ImageDataGenerator(
  rescale=1./255)

datagen_all = ImageDataGenerator( 
  rotation_range=20,
  width_shift_range=0.1,
  height_shift_range=0.1,
  shear_range=0.1,
  zoom_range=0.1,
  horizontal_flip=True,
  vertical_flip=True,
  rescale=1./255,                          
  )
```

```python
# 데이터 불러오기
# 훈련데이터와 테스트데이터 경로 지정
train_path='/content/drive/MyDrive/dataset/train'  
test_path='/content/drive/MyDrive/dataset/test'

# Generated train dataset
train_generator = datagen_train.flow_from_directory(train_path, 
                                       target_size=[224,224],
                                       color_mode='rgb', 
                                       class_mode="categorical", 
                                       batch_size = 32, 
                                       shuffle=True,
                                       subset='training') 

# Generated val dataset
validation_generator = datagen_train.flow_from_directory(train_path, 
                                       target_size=[224,224],
                                       color_mode='rgb', 
                                       class_mode="categorical",
                                       batch_size = 32,
                                       shuffle=True,
                                       subset='validation') 

# All train dataset
total_train_generator = datagen_all.flow_from_directory(train_path, 
                                       target_size=[224,224],
                                       color_mode='rgb', 
                                       class_mode="categorical",
                                       batch_size = 32,
                                       shuffle=True)

# test dataset 
test_generator = datagen_test.flow_from_directory(test_path, 
                                      target_size=[224,224],                                    
                                      color_mode='rgb', 
                                      class_mode="categorical",
                                      batch_size = 32) 
```

## CNN을 이용하여 딥러닝 모델 생성

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

model = keras.models.Sequential()

model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())

model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(6,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 20 에포크만큼 훈련
history = model.fit(train_generator,
          validation_data = validation_generator,
          steps_per_epoch = (train_generator.samples / train_generator.batch_size), 
          validation_steps = (validation_generator.samples / validation_generator.batch_size), 
          epochs=20)
```

```python
# Plotting Training and Validation Accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Acc','Val'], loc = 'upper left');
```

```python
# Plotting Training and Validation Error
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['loss','Val'], loc = 'upper left');
```

```python
# Evaluating the model on test data (bad loss result given)
result0 = model.evaluate(test_generator, steps = test_generator.samples/test_generator.batch_size)
print(result0)
print('Error on testing dataset: ', result0[0])
print('Accuracy on testing dataset: ', result0[1])
```

## VGG16을 이용하여 사전훈련된 딥러닝 모델 생성

```python
vgg16_model = keras.applications.VGG16(weights='imagenet', input_shape=(224, 224, 3), include_top=False) 
```

```python
vgg16_model.summary()
```

```python
vgg16_model.trainable = False # 모델 학습 중단
```

```python
# 사전훈련된 VGG16모델에 레이어 추가
# Create inputs with correct shape
inputs = vgg16_model.input

# Call the VGG16 model
x = vgg16_model(inputs,training=False)

# pooling layer 
x = keras.layers.GlobalAveragePooling2D()(x)

# Add final dense layer with softmax as activation function
outputs = keras.layers.Dense(6, activation = 'softmax')(x)

# Combine inputs and outputs to create model
model_vgg = keras.Model(inputs=inputs, outputs=outputs)
```

```python
model_vgg.summary()
```

```python
model_vgg.compile(loss='categorical_crossentropy',metrics=['accuracy'], optimizer='adam') 
```

```python
history_vgg = model_vgg.fit(train_generator,
          validation_data = validation_generator,
          steps_per_epoch = (train_generator.samples / train_generator.batch_size), 
          validation_steps = (validation_generator.samples / validation_generator.batch_size), 
          epochs=20)
```

```python
# Plotting Training and Validation Accuracy
plt.plot(history_vgg.history['accuracy'])
plt.plot(history_vgg.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Acc','Val'], loc = 'upper left')
```

```python
# Plotting Training and Validation Error
plt.plot(history_vgg.history['loss'])
plt.plot(history_vgg.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['loss','Val'], loc = 'upper left')
```

```python
# Evaluating the model on the testing dataset
result_vgg=model_vgg.evaluate(test_generator, steps=test_generator.samples/test_generator.batch_size)
print(result_vgg)
print('Error on testing dataset: ', result_vgg[0])
print('Accuracy on testing dataset: ', result_vgg[1])
```
